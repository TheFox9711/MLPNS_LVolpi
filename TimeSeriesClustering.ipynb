## Scaling (standardizing)

Standardize the data: We are working with time series, so each datapoint (=each country) has to be mean 0 and standard deviaion 1 along the rows (that is axis 1)

We use the `preprocessing.scale` function. Make sure you check! you want eavery country mean to be 0 and every country standard deviation to be 1...

````
# standardizing the data
X = pop_df_clean.values

# Per ogni riga normalizzo la popolazione, in modo che il valore medio della popolazione dal 1960 al 2020 sia 0 e la varianza 1.
pop_standard = preprocessing.scale(X, axis=1)
pop_standard
````

# Make a cluster model with KMeans
Use kmeans (`sklearn.cluster.KMeans`) to cluster the data. Make 9 models for number of clusters = 2, 3, 4, 5, 6, 7, 8, 9, 10. 

For each model, fit the data and calculate the "inertia" which is 

# Use an elbow plot to figure out ideal cluster _n_ 

````
# For n_clusters = 2 this would look as:
model = cluster.KMeans(n_clusters = 2, random_state=288605, n_init=10)
model.fit(pop_standard)
v = model.inertia_
print('The model inertia is:', f"{v:.4f}" )
````

````
# Now using the "inertia" (intracluster variance essentially) to make an Elbow plot: 
# inertia is the y axis, the number of clusters are the x axis.
# Now assessing if the elbow plot shows an inflection, which can be used to determine the "right" number of clusters
inertia = []
for i in range(2, 10):
  model = cluster.KMeans(n_clusters = i, random_state=288605, n_init=10)
  inertia.append(model.fit(pop_standard).inertia_)

pl.plot(range(2, 10), inertia)
pl.xlabel("Number of Clusters")
pl.ylabel("Inertia/Variance");
````

Figure 2: "Elbow plot" of k-means clustering of our time series. The y axis shows the sum of squared distances to the closest centroid. The point where inertia/variance dramatically drops in an elbow plot is recommended as the preferred number of clusters; this line has a bit too much of a gradual descent to make an effective recommendation.

````
model = cluster.KMeans(n_clusters = 4, random_state=288605, n_init=10)
model.fit(pop_standard)
v = model.inertia_
print('The model inertia is:', f"{v:.4f}" )
````

````
#Here we now chech the model lables
# Questi sono gli indici dei cluster a cui appartengono i dati, il numero i-esimo è l'indice del cluster a cui appartiene il dato
# i-esimo.
model.labels_
````

````
#Definting the models c, and determine the model centers in the k-means cluster
# Per ogni cluster ho l'andamento "medio" della popolazione, in questo caso il centroide di ogni cluster è come i suoi elementi
# un andamento demografico, quindi ognuno di essi è dato da una successione di N valori con N il numero di anni.
c = model.cluster_centers_
c
````

````
from matplotlib.pylab import float64
fig, ax = pl.subplots(2, 2, figsize=(8,10))

#... try to reproduce the plot below; one plot per cluster, with the centroid overplotted
for cl in range(4):
  axes = ax.flat[cl]
  axes.plot(pop_standard[model.labels_ == cl].T, 'k', alpha=0.2)
  axes.plot(c[cl,:], color='orange')
  axes.set_title('Cluster ' + str(cl+1))
  axes.set_xticks(range(0, 70, 10))
  axes.set_xticklabels(['%d'%i for i in range(1960, 2030, 10)])
````

Figure 3: Results of using k-means clustering on the standardized data seen in Figure 1. The mean line for each cluster is visible in orange.

Cluster 4 is a a fairly linear upward trend with some outliers and cluster 1 a more curved upward trend. Cluster 2 and cluster 3 have fewer time series; 2 rises steadily before a big drop in the 1990s whereas the countries in 3 kind of hit a plateau around the 1990s.

````
# For n_clusters = 4 this would look as:
# Un label -1 indica che il dato corrispondente è un outlier cioè non è stato possibile metterlo in alcun cluster.
model = cluster.DBSCAN(eps=0.2, min_samples=4)
model.fit(pop_standard)
model.labels_
````
